##  _quarto.xml에 페이지 구조설정
quarto에 내 홈페이지 어떤 모양을 띄게될지 설정하는 옵션이 있다.
홈페이지의 테마를 설정할수있고 [Bootswatch](https://quarto.org/docs/output-formats/html-themes.html)테마중 골라서 설정한다.

```xml
format:
  html:
    theme: darkly
```

[Navigation설정](https://quarto.org/docs/websites/website-navigation.html)을 참고하여 네비게이션 바를 만들자
사이드바 네비바 다양하게 제공한다. 자동 생성도 해주지만 중구난방으로 나오기 때문에 하나씩 설정해주자

```
website:
  sidebar:
    style: docked
    search: true
    contents:
      - section: Quarto
        contents:
          - "python-notebook/quarto/1.Start-Quarto.ipynb"
          - "python-notebook/quarto/2.QuartoWithGitPage.ipynb"
          - "python-notebook/quarto/3.Quarto-CI-GitActions.ipynb"
          - "python-notebook/quarto/4.Quarto-website.ipynb"
```


## index.qmd를 작성한다
머리글을 쓸게 없는데 메인페이지를 뭘로 해야 하나 했는데
컨텐츠 리스팅을 해줄수 있어서 적용하였다
[가이드](https://quarto.org/docs/websites/website-listings.html)


```
---
title: "Pages"
listing: 
  type: grid
  contents:
    - "python-notebook/quarto/*.ipynb"
  sort: "filename"
---
```

리스팅은 Sort기능이 있다... 네비게이션도 언젠가는 지원해 줄지도 모르겠다.
